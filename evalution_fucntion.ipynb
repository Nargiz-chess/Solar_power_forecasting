```python
import matplotlib.pyplot as plt

def evaluation(actual, predicted):
  error = abs(actual-predicted)
  total_error = sum(error)
  error_percentage = np.round((error/total_error) * 100, decimals=0)
  return error_percentage.astype(int)

def plot (array2):
  list_percentage = []
  for val in array2:
    item = str(val)+"%"
    list_percentage.append(item)
  hours = [h for h in range(0,len(array2))]
  plt.plot(hours,list_percentage,marker='o');plt.xlabel('Hours');plt.ylabel('Error in %');plt.title('Percentage error per hour')
  return plt.show()
```

```python
import numpy as np

actual = np.array([1,-2,3,2,5,0,-1])
predicted = np.array([0,1,3,4,5,-1,-2])

errors = evaluation(actual,predicted)

plot(errors)
```
